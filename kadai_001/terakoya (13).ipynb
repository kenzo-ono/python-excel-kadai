```python
# openpyxl.stylesモジュールのPatternFillクラスをインポート
from openpyxl.styles import PatternFill

# ワークブックを作成
workbook = openpyxl.Workbook()

# アクティブなシートを変数「sheet」に格納
sheet = workbook.active

# A1セルの背景色を赤色（#FF0000）に設定
sheet['A1'].fill = PatternFill(patternType='solid', fgColor='FF0000')

workbook.save('sample4.xlsx')
```

```python
# openpyxl.stylesモジュールのFontクラスをインポート
from openpyxl.styles import Font

# ワークブックを作成
workbook = openpyxl.Workbook()

# アクティブなシートを変数「ws」に格納
ws = workbook.active

# フォント名、サイズ、太字の設定
font = Font(name='Calibri', size=30, bold=True)

# A1セルに文字列を入力
ws['A1'] = 'Hello, World!'

# A1セルにフォントを適用
ws['A1'].font = font

# ファイルを別名保存
workbook.save('sample5.xlsx')
```

```python
# openpyxl.stylesモジュールのBorderクラスとSideクラスをインポート
from openpyxl.styles import Border, Side

# 新規ワークブックを作成
workbook = openpyxl.Workbook()

# アクティブなシートを取得し変数wsに格納
ws = workbook.active

# B2セルに文字列データを入力
ws['B2'] = 'Hello World!'

# Side属性で線のスタイルを指定
thin_border = Side(style='thin')

# Border属性で各辺の線のスタイルを指定
border = Border(top=thin_border, bottom=thin_border, left=thin_border, right=thin_border)

# B2のborder属性にborderを設定
ws['B2'].border = border

# ワークブックを保存
workbook.save('sample6.xlsx')
```

```python
from openpyxl import Workbook
from datetime import datetime

# 新しいExcelファイルを作る
wb = Workbook()
ws = wb.active

# 今日の日付を取得（YYYY/MM/DD形式）
today = datetime.today().strftime("%Y/%m/%d")

# タイトル
ws["B2"] = "請求書"

# 会社情報
ws["B4"] = "株式会社ABC"
ws["B5"] = "〒101-0022 東京都千代田区神田練塀町300"
ws["B6"] = "TEL:03-1234-5678 FAX:03-1234-5678"
ws["B7"] = "担当者名:鈴木一郎 様"

# 請求番号と日付
ws["F4"] = "No."
ws["G4"] = "0001"
ws["F5"] = "日付"
ws["G5"] = today

# 見出し
ws["B10"] = "商品名"
ws["C10"] = "数量"
ws["D10"] = "単価"
ws["E10"] = "金額"

# 商品データ
ws["B11"] = "商品A"
ws["C11"] = 2
ws["D11"] = 10000
ws["E11"] = 20000

ws["B12"] = "商品B"
ws["C12"] = 1
ws["D12"] = 15000
ws["E12"] = 15000

# 小計・税・合計
ws["B15"] = "小計"
ws["E15"] = 35000

ws["B16"] = "消費税"
ws["E16"] = 3500

ws["B17"] = "合計"
ws["E17"] = 38500

# ファイル名（今日の日付つき）
filename = f"請求書_{today.replace('/', '')}.xlsx"

# 保存
wb.save(filename)

print(f"{filename} を作成しました！")
```
