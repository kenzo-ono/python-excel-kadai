```python
import pandas as pd

df_2022 = pd.read_excel("2022_年間売上表.xlsx")
df_2023 = pd.read_excel("2023_年間売上表.xlsx")

print(df_2022.head())
print(df_2023.head())
```

```python
df_all = pd.concat([df_2022, df_2023])
print(df_all)
```

```python
summary = (
    df_all
    .groupby(["商品", "売上年"], as_index=False)["金額（千円）"]
    .sum()
)

print(summary)
```

```python
summary.to_excel("売上集計表.xlsx", index=False)
```

```python
from openpyxl import load_workbook
from openpyxl.styles import PatternFill

wb = load_workbook("売上集計表.xlsx")
ws = wb.active

gray = PatternFill(fill_type="solid", fgColor="F2F2F2")

for cell in ws[1]:
    cell.fill = gray

wb.save("売上集計表.xlsx")
```
